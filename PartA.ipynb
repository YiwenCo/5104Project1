```python
import pandas as pd

df = pd.read_csv("Concrete_Data.csv")

print(df.head())
```

```python
# change the column name for coding
df.columns = [
    "Cement",
    "Slag",
    "FlyAsh",
    "Water",
    "Superplasticizer",
    "CoarseAgg",
    "FineAgg",
    "Age",
    "Strength"
]

df.describe()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
corr_matrix = df.corr()
print(corr_matrix)
```

```python
corr_strength = corr_matrix["Strength"].sort_values(ascending=False)
print(corr_strength)
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=False, cmap="coolwarm")
plt.show()
```
